```python
import requests
import bs4
import re
```

```python
response = requests.get('https://gb.youtubers.me/thailand/all/top-1000-most-subscribed-youtube-channels-in-thailand')
```

```python
response.encoding = 'utf-8'
```

```python
soup = bs4.BeautifulSoup(response.text, 'html.parser')
```

```python
soup.title.string
```

```python
table = soup.find('table', class_='top-charts')
```

```python
import pandas as pd
```

```python
# Defining of the dataframe
df = pd.DataFrame(columns=['rank', 'youtuber', 'subscribers', 'video views', 'video count', 'category', 'started'])

# Collecting Ddata
for row in soup.findAll('table')[0].findAll('tr'):    
    # Find all data for each column
    columns = row.find_all('td')

    if(columns != []):    
        #print(columns)
        rank = columns[0].text.strip()
        youtuber = columns[1].text.strip()
        subscribers = columns[2].text.strip()
        video_views = columns[3].text.strip()
        video_count = columns[4].text.strip()
        category = columns[5].text.strip()
        started = columns[6].text.strip()
        #print(rank)
        #print(youtuber)
        #print(subscribers)
        #print(video_views)
        #print(video_count)
        #print(category)
        #print(started)

        df = df.append({'rank': rank,  'youtuber': youtuber, 'subscribers': subscribers, 'video views': video_views, 'video count': video_count, 'category': category, 'started': started}, ignore_index=True)
```

# Top 1000 Video

```python
# -*- coding: utf8 -*-
import requests
import bs4
```

```python
response_1 = requests.get('https://gb.youtubers.me/thailand/all/top-1000-youtube-videos-in-thailand')
```

```python
response_1.encoding = 'utf-8'
```

```python
soup = bs4.BeautifulSoup(response_1.text, 'html.parser')
```

```python
soup.title.string
```

```python
table = soup.find('table', class_='top-charts').encode()
```

```python
# Defining of the dataframe
df = pd.DataFrame(columns=['rank', 'video', 'video views', 'likes', 'dislikes', 'category','published'])

# Collecting Ddata
for row in soup.findAll('table')[0].findAll('tr'):    
    # Find all data for each column
    columns = row.find_all('td')

    if(columns != []):    
        #print(columns)
        rank = columns[0].text.strip()
        video = columns[1].text.strip()
        #print(video)
        video_views = columns[2].text.strip()
        likes = columns[3].text.strip()
        dislikes = columns[4].text.strip()
        category = columns[5].text.strip()
        published = columns[6].text.strip()

        df = df.append({'rank': rank,  'video': video, 'video views': video_views, 'likes': likes, 'dislikes': dislikes, 'category': category, 'published': published}, ignore_index=True)
```

```python
df.head()
```

```python
df.to_excel("G2 - Top_1000_Videos (update 30.03.2023).xlsx")
```
